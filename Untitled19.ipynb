```python
mean_age = np.mean(data['Age'])
print("Mean Age of Passengers is - ",mean_age)
```

```python
mean_fare = np.mean(data['Fare'])
print("Mean Fare is - ",mean_fare)
```

```python
meadian_age = np.median(data['Age'])
print("Median value of Age -", meadian_age)
```

```python
median_fare = np.median(data['Fare'])
print("Median value of Fare -", median_fare)
```

```python
mode_age = data['Age'].mode()
print("Mode value of Age -", mode_age)
```

```python
mode_class = data['Pclass'].mode()
print("Mode value of Class -", mode_class)
```

```python
mode_gender = data['Gender'].mode()
print("Mode value of Gender -", mode_gender)
```
