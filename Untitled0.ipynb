```python
def printnumber(n):
    iteration=0
    print("The number entered by the user is ",n)
    iteration+=1
    print("Total iterations done by code is ",iteration,"\n")

printnumber(10)
printnumber(20)
```

```python
def Ontime(n):
  iteration=0
  for i in range(1,n+1):
    iteration+=1
  print("When n is ",n," Iterations = ",iteration)
Ontime(10)
Ontime(20)
```

```python
def ONSquareTime(n):
  iteration=0
  for i in range(0,n):
    for j in range(0,n):
      print("*", end =" ")
      iteration+=1
    print("")
  print("\nWhen n is ",n," Iterations = ",iteration,"\n")

ONSquareTime(10)
ONSquareTime(20)
```
